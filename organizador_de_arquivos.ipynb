```python
import os

files_list = []

types = []

palavra1 = ''

tipo = ''

cwd = os.getcwd()

full_list = os.listdir(cwd)

#buscando arquivos na pasta e verificando se o arquivo é uma pasta ou .py

for arquive_name in full_list:

  if os.path.isfile(arquive_name) and ',py' not in arquive_name:

    files_list.append(arquive_name)

#Separando extensões dos arquivos e colocando na lista de extensões

for archive_name in files_list:

  for index, c in enumerate(archive_name,1):

    word = archive_name[index*(-1)]

    palavra1 += word

    if word == '.':

      for index, d in enumerate(palavra1,1):

        word = palavra1[index*(-1)]

        tipo += word

        if index == len(palavra1):

          types.append(tipo.lower())

          tipo = ''

      palavra1 = ''

      break

types = set(types)

#Criando as pastas

for file_type in types:

  try:

    os.mkdir(f'Arquivos {file_type}')

  except:

    print(f'Pasta [Arquivos {file_type}] já criada')

#Inserindo os arquivos nas pastas criadas

for file_type in types:

  for index, archive_name in enumerate (files_list):

    if file_type in archive_name:

      from_path = os.path.join(cwd,archive_name)

      to_path = os.path.join(cwd,f'Arquivos {file_type}',archive_name)

      os.replace(from_path, to_path)
```
